```python
import pandas as pd

meeting_note_csv_path = "./audio/guitar_final.csv"
df_rttm = pd.read_csv(meeting_note_csv_path, sep="|")

display(df_rttm)
```

```python
name_dict = {
  "SPEAKER_00": "AI",
  "SPEAKER_01": "박성철",
}

df_rttm["name"] = df_rttm["speaker_id"].apply(lambda x: name_dict[x])

display(df_rttm)
```

```python
meeting_note_text = df_rttm[['start', 'end', 'name', 'text']].to_json(orient='records', force_ascii=False)
print(meeting_note_text)
```

```python
system_prompt = f"""
  너는 회의 내용을 요약해주는 봇이다. 아래 회의록을 읽고, 주요 내용을 요약해주고, 회의록에 모호한 내용이 있으면 수정해줘.
  결과는 마크다운 형식으로 작성한다.
  
  아래 형식에 맞추어 작성하라.
  
  # 회의 제목 
  
  ## 주요 내용 
  
  ## 참석자별 입장
  
  ## 결정사항
  
  ========= 이하 회의록 =========
  
  {meeting_note_text}
"""

from openai import OpenAI
from dotenv import load_dotenv
import os 

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

response = client.chat.completions.create(
  model="gpt-5-mini",
  messages=[{"role": "system", "content": system_prompt}]
)

summary = response.choices[0].message.content
summary = summary.strip()

print(summary)

with open("summary.md", "w") as f:
  f.write(summary)
```
