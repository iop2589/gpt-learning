```python
from langchain_community.adapters.openai import convert_openai_messages 
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

query = "현대자동차 미국 시장 2026년 전망"

content = client.search(
  query=query,
  search_depth="advanced",
  include_raw_content=True,
)["results"]
```

```python
prompt = [{
  "role": "system",
  "content": (
    "당신은 신문 기사를 쓰는 기자 AI입니다. \n"
    "당신은 주어진 정보를 바탕으로 객관적이고 체계적으로 작성된 기사를 써야합니다. \n" 
  )
}, {
  "role": "user",
  "content": (
    f'정보: """{content}"""\n\n'
    f'위의 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요: "{query}"\n'
    '-신문 기사 형식을 사용하되, MLA 표준을 준수하는 markdown 문법으로 작성하세요.'
    '-활용한 자료는 출처를 명시하세요.'
  )
}]
```

```python
report = llm.invoke(prompt).content

print(report)
```
